As per the analysis from above data, we inferred that majority of transactions has a finality time of below 20 seconds. There are only 13 transactions who crosses the time of above 20 seconds out of 181 transactions.So, overall standard deviation is not high and mean value is 16.8. And hence, the finality time approximate value is 120sec. (Mean + std deviation + considerable transactions having time above 30 seconds + outliers consideration.)
